```python
dbutils.widgets.text("file_name","")
```

```python
p_file_name = dbutils.widgets.get("file_name")
```

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "parquet")\
    .option(
             "cloudFiles.schemaLocation",
             f"abfss://bronze@databricksetestorage2003.dfs.core.windows.net/checkpoints_{p_file_name}"
    )\
    .load(
             f"abfss://source@databricksetestorage2003.dfs.core.windows.net/{p_file_name}"
    )
```

```python
df.writeStream.format("parquet")\
    .option(
        "checkpointLocation",
        f"abfss://bronze@databricksetestorage2003.dfs.core.windows.net/checkpoints/{p_file_name}"
    )\
    .option(
        "path",
        f"abfss://bronze@databricksetestorage2003.dfs.core.windows.net/data/{p_file_name}"
    )\
    .trigger(once=True)\
    .start()
```
