```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("parquet")\
               .load("abfss://bronze@databricksetestorage2003.dfs.core.windows.net/data/customers")
```

```python
display(df)
```

```python
df = df.drop("_rescued_data")
```

```python
df = df.withColumn('domains', split(col('email'), '@')[1])
```

```python
display(df)
```

```python
display(
    df.groupBy('domains')
      .agg(count("customer_id").alias("total_customers"))
      .sort('total_customers', ascending=False)
)
```

```python
df_gmail = df.filter(col("domains") == 'gmail.com')

df_gmail = df.filter(col("domains") == 'gmail.com')


display(
    df_gmail.groupBy('domains')
            .agg(count("customer_id").alias("total_customers"))
)
```

```python
df = df.withColumn('full_name', concat(col('first_name'), lit(' '),col('last_name')))

display(df)
```

```python
df.write.mode("append").format("delta").save("abfss://silver@databricksetestorage2003.dfs.core.windows.net/customers")
```
